```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('Telco-Customer-Churn.xlsx')
data
```

```python
data.isna().any()
```

**Compare churn count with respect to gender**

```python
plt.figure(figsize = (10,8))
sns.countplot(data["gender"],hue = data["Churn"])
```

**Find out how many female senior citizens there in the dataset**

```python
plt.figure(figsize = (10,8))
sns.countplot(data["gender"],hue = data["SeniorCitizen"])
```

**Compare 'tenure' with 'Total Charges'**

```python
plt.figure(figsize = (10,8))
plt.scatter(data.tenure,data.TotalCharges,label="Total Charges")
plt.xlabel("tenure")
plt.ylabel("total Charges")
```

**Find out which contract is preferred by the senior citizen**

```python
sns.countplot(data["Contract"],hue = data["SeniorCitizen"])
```
